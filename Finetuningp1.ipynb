```python
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig,HfArgumentParser,TrainingArguments,pipeline, logging
from peft import LoraConfig, PeftModel, prepare_model_for_kbit_training, get_peft_model
import os,torch, wandb
from datasets import load_dataset
from trl import SFTTrainer
```

```python
base_model="mistralai/Mistral-7B-v0.1"
```

```python
from transformers import AutoModelForCausalLM, BitsAndBytesConfig
import torch

bnb_config = BitsAndBytesConfig(
    load_in_4bit= True,
    bnb_4bit_quant_type= "nf4",
    bnb_4bit_compute_dtype= torch.bfloat16,
    bnb_4bit_use_double_quant= False,
)
model = AutoModelForCausalLM.from_pretrained(
        base_model,
        #load_in_4bit=True,
        quantization_config=bnb_config,
        torch_dtype=torch.bfloat16,
        device_map="auto",
        trust_remote_code=True,
)
model.config.use_cache = False # silence the warnings
model.config.pretraining_tp = 1
model.gradient_checkpointing_enable()


```

```python
tokenizer = AutoTokenizer.from_pretrained("mistralai/Mistral-7B-v0.1", trust_remote_code=True)
tokenizer.padding_side = 'right'
tokenizer.pad_token = tokenizer.eos_token
tokenizer.add_eos_token = True
tokenizer.add_bos_token, tokenizer.add_eos_token
```

```python
model = prepare_model_for_kbit_training(model)
peft_config = LoraConfig(
    lora_alpha=16,
    lora_dropout=0.1,
    r=64,
    bias="none",
    task_type="CAUSAL_LM",
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj","gate_proj"]
)
model = get_peft_model(model, peft_config)
```

```python
training_arguments = TrainingArguments(
    output_dir="./results",
    num_train_epochs=1,
    per_device_train_batch_size=4,
    gradient_accumulation_steps=1,
    optim="paged_adamw_32bit",
    save_steps=25,
    logging_steps=25,
    learning_rate=2e-4,
    weight_decay=0.001,
    fp16=False,
    bf16=False,
    max_grad_norm=0.3,
    max_steps=-1,
    warmup_ratio=0.03,
    group_by_length=True,
    lr_scheduler_type="constant",
    report_to="wandb"
)
```

```python
import re

def clean_text(text):
    sections_to_remove = [
        "Notable people",
        "References",
        "External links",
        "See also",
        "History",
        "Research"
    ]


    pattern = r'(' + '|'.join(re.escape(section) for section in sections_to_remove) + r')\s*[\n\r].*'


    cleaned_text = re.split(pattern, text, flags=re.IGNORECASE)[0]

    return cleaned_text.strip()
```

```python
for index, row in df.iterrows():
    raw_text = row["page_text"]
    cleaned = clean_text(raw_text)
    print(cleaned)
    print("==== ligne nettoyée ====")

    if index == 9:
        break

```

```python
df["clean_text"] = df["page_text"].apply(clean_text)
```

```python
import json

# Création du fichier JSONL
with open("medical_conditions_finetune.jsonl", "w", encoding="utf-8") as f:
    for _, row in df.iterrows():
        instruction = "What is " + row["page_title"] + "?"
        input_text = ""
        output = row["clean_text"]

        item = {
            "instruction": instruction,
            "input": input_text,
            "output": output.strip()
        }

        f.write(json.dumps(item, ensure_ascii=False) + "\n")

print(" Fichier JSONL prêt pour le fine-tuning.")
```

```python
import json

# Charger ton ancien fichier
with open("medical_conditions_finetune.jsonl", "r", encoding="utf-8") as old_file, \
     open("medical_causal_format.jsonl", "w", encoding="utf-8") as new_file:

    for line in old_file:
        item = json.loads(line)

        instruction = item["instruction"]
        input_text = item["input"]
        output = item["output"]

        if input_text:
            prompt = f"### Instruction:\n{instruction}\n\n### Input:\n{input_text}\n\n### Response:\n{output}"
        else:
            prompt = f"### Instruction:\n{instruction}\n\n### Response:\n{output}"

        new_item = {"text": prompt}
        new_file.write(json.dumps(new_item, ensure_ascii=False) + "\n")

print("Fichier causal prêt pour le fine-tuning.")
```
